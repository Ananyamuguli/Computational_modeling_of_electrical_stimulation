```python
import gc
import pickle as pkl
import numpy as np
import matplotlib.pyplot as plt
import glob
import PyDSTool as dst
```

# Glutamate Traces

```python
## Files
nostim_traces = pkl.load(open('MOTIF1/glutamate_params_nostim_traces_2.pkl', 'rb'))
acstim_traces = pkl.load(open('MOTIF1/glutamate_params_acstim_traces_2.pkl', 'rb'))
```

## No stimulation

```python
glu_vals = [g for g in nostim_traces.keys()]
print(glu_vals)
```

```python
gvals_paper = [0,0.2]
upper_limit = [20000, 39500]
fig, axes = plt.subplots(2,1, figsize=(10, 6))
for g,ul, ax in zip(gvals_paper,upper_limit, axes):
    ode, traj, pts = nostim_traces[g]
    ax.plot(pts['t'][1000:ul], pts['Vs'][1000:ul], color='blue', alpha = 0.5, label = 'PYR1')
    ax.plot(pts['t'][1000:ul], pts['Vpost'][1000:ul], color = 'green', alpha=0.6, label='PYR2')
    
    ax.tick_params(which='both', labelsize=13)
    ax.set_ylabel('Membrane \n potential (mV)', fontsize=14)
    ax.set_xlabel('Time (ms)', fontsize=14)
    ax.set_title('{} mM'.format(g), fontsize=17)
    
plt.tight_layout()
for ax in axes.flat:
    h, l = ax.get_legend_handles_labels()
fig.legend(h,l, bbox_to_anchor=(1.0, 0.05), fontsize=15)
plt.savefig('Results/motif1_glutamate_traces_appendix_nostim.pdf',  bbox_inches='tight')
plt.savefig('Results/motif1_glutamate_traces_appendix_nostim.png',  bbox_inches='tight')

```

```python
for g in nostim_traces.keys():
    fig, ax = plt.subplots(figsize=(25, 5))
    ode, traj, pts = nostim_traces[g]
    ax.plot(pts['t'], pts['Vs'], color='blue', alpha = 0.6, label = 'PYR1')
    ax.plot(pts['t'], pts['Vpost'], color = 'red', alpha=0.6, label='PYR2')
    ax.legend(loc='best')
    ax.tick_params(which='both', labelsize=13)
    ax.set_ylabel('Membrane potential (mV)', fontsize=15)
    ax.set_xlabel('Time (ms)', fontsize=15)
    plt.title('{} mM'.format(g), fontsize=20)
    plt.savefig('Results/motif1_glutamate{}_traces_nostim.pdf'.format(g))
```

## Stimulation

```python
glu_vals = [g for g in acstim_traces.keys()]
print(glu_vals)
#First stimulation point
stimparams = [params for params in acstim_traces[glu_vals[0]].keys()]
print(stimparams)
```

```python
upper_limit = [31000, 26000, 42000, 76000]
fig, axes = plt.subplots(4,1, figsize=(15, 12))
acstim_vals = [(key, val) for key, val in acstim_traces[glu_vals[0]].items()]
for stim, ul, ax in zip(acstim_vals, upper_limit, axes):
    ode, traj, pts = stim[1]
    ax.plot(pts['t'][1000:ul], pts['Vs'][1000:ul], color='blue', alpha = 0.4, label = 'PYR1')
    ax.plot(pts['t'][1000:ul], pts['Vpost'][1000:ul], color = 'green', alpha=0.4, label='PYR2')
    
    ax.tick_params(which='both', labelsize=13)
    ax.set_ylabel('Membrane \n potential (mV)', fontsize=14)
    ax.set_xlabel('Time (ms)', fontsize=14)
    ax.set_title('{} Hz, {} $\mu A/cm^2$'.format(stim[0][0], stim[0][1]), fontsize=17)
    
plt.tight_layout()
for ax in axes.flat:
    h, l = ax.get_legend_handles_labels()
fig.legend(h,l, bbox_to_anchor=(1.0, 0.05), fontsize=15)
plt.savefig('Results/motif1_glutamate_traces_appendix_acstim_stim_analysis_plots.pdf',  bbox_inches='tight')
plt.savefig('Results/motif1_glutamate_traces_appendix_acstim_stim_analysis_plots.png',  bbox_inches='tight')
```
